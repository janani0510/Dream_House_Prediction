```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import joblib

df = pd.read_csv("homeprices.csv")
x = df[["area"]]
y = df['price']

model = LinearRegression()
model.fit(x,y)

print("weight(coef_):",model.coef_)
print("bias(intercept_):",model.intercept_)

joblib.dump(model,"Dream_House_Price.joblib")
print("MODEL SAVED SUCCESSFULLY!")

loaded_model = joblib.load("Dream_House_Price.joblib")

area_value = 5000
input_data = pd.DataFrame([[area_value]],columns =['area'])
prediction = loaded_model.predict(input_data)
print(f"predicted price for{area_value}sqft:",prediction[0])
```
